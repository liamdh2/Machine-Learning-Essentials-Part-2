## Association Rule Mining Using Apriori Algorithm

```python
%pip install mlxtend
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('/content/drive/MyDrive/python-for-data-science-and-machine-learning-essential-training-part-1-3006708-main (1)/python-for-data-science-and-machine-learning-essential-training-part-2-3016934-main/data/groceries.csv')
```

```python
data.head()
```

```python
basket_sets = data.stack().str.get_dummies().groupby(level=0).sum()
```

```python
basket_sets.head()
```

```python
basket_sets_bool = basket_sets.astype(bool)
apriori(basket_sets_bool, min_support=0.02, use_colnames=True)
```

```python
df_combo = basket_sets.astype(bool)
freq_itemsets = apriori(df_combo, min_support=0.002, use_colnames=True)
freq_itemsets['length'] = freq_itemsets['itemsets'].apply(lambda x: len(x))
freq_itemsets
```

```python
freq_itemsets[(freq_itemsets['length']>=3)]
```

```python
rule = association_rules(freq_itemsets, metric='confidence', min_threshold=0.9)
rule.head()
```

```python
rule = association_rules(freq_itemsets, metric='lift', min_threshold=1)
rule.head()
```

```python
rule[(rule['lift']>=5 &(rule['confidence']>=.8))]
```
