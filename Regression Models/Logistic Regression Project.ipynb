**Data Preparation**

```python
sns.countplot(x='Survived', hue= 'Survived', data = titanic_training, palette='hls', legend= False)
```

```python
titanic_training.isnull().sum()
```

```python
titanic_training.describe()
```

```python
titanic_data = titanic_training.drop(['Name', 'Ticket', 'Cabin'], axis=1)
titanic_data
```

**Treat Missing Values**

```python
sns.boxplot(x='Parch', y='Age', data=titanic_data)
```

```python
Parch_groups = titanic_data.groupby(titanic_data['Parch'])
Parch_groups.mean(numeric_only=True)
```

```python
def age_approx(cols):
  Age = cols.iloc[0]
  Parch = cols.iloc[1]

  #if null where the Parch = i, replace Parch with average at i
  if pd.isnull(Age):
    if Parch == 0:
      return 32
    elif Parch == 1:
      return 24
    elif Parch == 2:
      return 17
    elif Parch == 3:
      return 33
    elif Parch == 4:
      return 45
    else:
      return 30
  else:
    return Age
```

```python
titanic_data['Age'] = titanic_data[['Age','Parch']].apply(age_approx,axis=1)
titanic_data.isnull().sum()
```

```python
#last null values are under Embarked and there are only 2 therefore insignificant if records are dropped
titanic_data.dropna(inplace=True)
titanic_data.reset_index(inplace = True, drop=True)
print(titanic_data.info())
```

**Re-Encode Variables**

```python
#Encode classifier variables such as sex into numeric classifiers
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
gender_cat = titanic_data['Sex']
gender_encoded = label_encoder.fit_transform(gender_cat)
gender_encoded[0:5]
```

```python
titanic_data.head()
```

```python
gender_df = pd.DataFrame(gender_encoded, columns=['male_gender'])
gender_df.head()
```

```python
embarked_cat = titanic_data['Embarked']
embarked_encoded = label_encoder.fit_transform(embarked_cat)
embarked_encoded[0:100]
```

```python
#Use one hot encoder for Embarked
from sklearn.preprocessing import OneHotEncoder
binary_encoder = OneHotEncoder(categories='auto')
embarked_1hot = binary_encoder.fit_transform(embarked_encoded.reshape(-1,1))
embarked_1hot_mat = embarked_1hot.toarray()
embarked_df = pd.DataFrame(embarked_1hot_mat, columns = ['C', 'Q', 'S'])
embarked_df.head()
```

```python
titanic_data.drop(['Sex', 'Embarked'], axis=1, inplace=True)
titanic_data.head()
```

```python
titanic_dmy = pd.concat([titanic_data, gender_df, embarked_df], axis=1, verify_integrity=True).astype(float)
titanic_dmy[0:5]
```

**Data Validation**

```python
sns.heatmap(titanic_dmy.corr())
```

```python
#fair and Pclass are not independent
titanic_dmy.drop(['Fare'], axis=1, inplace=True)
titanic_dmy.head()
```

```python
titanic_dmy.info()
```

**Model Deployment**

```python
X_train, X_test, y_train, y_test= train_test_split(titanic_dmy.drop('Survived', axis=1),titanic_dmy['Survived'], test_size = 0.20, random_state = 200)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_train[0:5]
```

```python
LogReg= LogisticRegression(solver='liblinear')
LogReg.fit(X_train.values, y_train.values)
```

```python
y_pred = LogReg.predict(X_test.values)
```

**Model Evaluation**

```python
print(classification_report(y_test, y_pred))
```

```python
y_train_pred = cross_val_predict(LogReg, X_train, y_train, cv=5)
confusion_matrix(y_train, y_train_pred)
```

```python
precision_score(y_train, y_train_pred)
```
